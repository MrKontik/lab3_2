```python
def simuv(data,userInd,v):
    u=data[userInd:userInd+1].values[0][1:]
    v=data[v-1:v].values[0][1:]
    sumUV=0
    sumUU=0
    sumVV=0
    for i in range(0,30):
        if (u[i]!=-1 and v[i]!=-1):
            sumUV+=u[i]*v[i]
            sumUU+=u[i]*u[i]
            sumVV+=v[i]*v[i]
    return sumUV/math.sqrt(sumUU*sumVV)
```

```python
def ru(data,userInd):
    u=data[userInd:userInd+1].values[0][1:]
    sumF=0
    scope=0
    for i in range(0,30):
        if (u[i]!=-1):
            scope+=1
            sumF+=u[i]
    return sumF/scope
```

```python
def kNN(data,userInd,k):
    res=np.array([[0,0]])
    u=data[userInd:userInd+1].values[0][1:]
    for i in range(0,40):
        if (i!=userInd):
            res=np.append(res, [[i,simuv(data,userInd,i+1)]], axis = 0)
        else:
            res=np.append(res, [[i,0]], axis = 0)
    res=res[1:]
    arr=res
    arr=arr[arr[:, 1].argsort()]
    arr=arr[40-k:]
    return arr
```

```python
def rui(data,userInd,k,i,friend):
    sum_sim=0
    sum_abs_sim=0
    u=data[userInd:userInd+1].values[0][1:]
    for g in range(1,k+1):
        v=data[friend[g-1]:friend[g-1]+1].values[0][1:]
        if (v[i]!=-1):
            sum_sim=sum_sim+simuv(data,userInd,friend[g-1]+1)*(v[i]-ru(data,friend[g-1]))
            sum_abs_sim=sum_abs_sim+np.absolute(simuv(data,userInd,friend[g-1]+1))
    return ru(data,userInd)+sum_sim/sum_abs_sim
```

```python
#это и есть ответ для первой части задания
def task1(user,k):
    res1=[]
    u=data[user-1:user].values[0][1:]
    knn=kNN(data,user-1,k)
    friend=np.array([knn[0][0],knn[1][0],knn[2][0],knn[3][0]])
    friend=friend.astype(int)
    for l in range(0,30):
        if (u[l]==-1):
            res1=res1+[l+1]
            res1=res1+[np.around(rui(data,user-1,4,l,friend),decimals = 3)]
    print(res1)
    return res1
```

```python
sparql = SPARQLWrapper("https://query.wikidata.org/sparql")

sparql_query = """

SELECT ?item ?itemLabel ?actor ?actorLabel
WHERE 
{  
  ?item wdt:P1441 wd:Q171669.
  ?item wdt:P175 ?actor.
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}
 
"""

sparql.setQuery(sparql_query)

sparql.setReturnFormat(JSON)
results = sparql.query().convert()
```
